```python
import json
import numpy as np

for parte in all:
  parte["TIENDA"] = parte["TIENDA"].str.strip()
  parte["CATEGORIA"] = parte["CATEGORIA"].str.strip()
  parte["DESCRIPCION"] = parte["DESCRIPCION"].str.strip()
  parte["DETALLE"] = parte["DETALLE"].str.strip()
  parte["SALDO"] = parte["SALDO"].apply(lambda x: 0 if isinstance(x, str) else x)
  parte["SALDO"] = parte["SALDO"].fillna(0)
  parte["%"] = parte["%"].apply(lambda x: None if isinstance(x, str) else x)
  parte["PRECIO"] = parte["PRECIO"].apply(lambda x: None if isinstance(x, str) else x)
  parte["PRECIO OFERTA"] = parte["PRECIO OFERTA"].apply(lambda x: None if isinstance(x, str) else x)

# Unir todas las partes en un solo DataFrame
df_completo = pd.concat(all, ignore_index=True)

# Reemplazar NaN con None
df_completo = df_completo.replace({np.nan: None})

# Convertir el DataFrame a un array de diccionarios
productos = df_completo.to_dict(orient='records')

# Convertir el array de productos a JSON
productos_json = json.dumps(productos, ensure_ascii=False, indent=4)

# Imprimir o guardar el JSON resultante
productos_json
```
